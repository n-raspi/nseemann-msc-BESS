```python
import pandas as pd
pd.options.plotting.backend = "plotly"
```

```python
load_20 = pd.read_csv('../../20_data/load_profiles_braeuer/LoadProfile_20IPs_2016.csv', delimiter = ';', skiprows=1)
load_30 = pd.read_csv('../../20_data/load_profiles_braeuer/LoadProfile_30IPs_2017.csv', delimiter = ';', skiprows=1)
```

```python
load_20.dropna(how='any', inplace = True)
load_20['Time stamp'] = load_20['Time stamp'].apply(lambda x: x.strip('b '))
load_20['Time stamp'] = pd.to_datetime(load_20['Time stamp'], format = "%d.%m.%Y %H:%M:%S")
load_20.set_index('Time stamp', inplace = True)
load_20['LG 11'].plot()
```
